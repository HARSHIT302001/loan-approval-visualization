```python
# Fill missing values
for col in df.columns:
    if df[col].dtype == "object":
        df[col].fillna(df[col].mode()[0], inplace=True)  # Fill categorical columns with mode
    else:      
        df[col].fillna(df[col].mean(), inplace=True)  # Fill numeric columns with mean

# Check if missing values are handled
print("Missing values after handling:\n", df.isnull().sum())
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("loan_sanction_test.csv")  # Change path if needed

# Check current data types
print("Before conversion:\n", df.dtypes)

# Convert Loan_Amount_Term and Credit_History to integer type
df["Loan_Amount_Term"] = df["Loan_Amount_Term"].astype("Int64")  # Allows NaN values
df["Credit_History"] = df["Credit_History"].astype("Int64")  # Allows NaN values

# Check updated data types
print("\nAfter conversion:\n", df.dtypes)
```

## UNIVARIATE ANALYSIS

### DISTRIBUTION OF NUMERICAL COLUMNS 

```python

# List of numerical columns
numeric_columns = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']

# Histogram & KDE Plot
plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col].dropna(), bins=30, kde=True, color="blue")
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# Box Plots
plt.figure(figsize=(12, 6))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x=df[col], color='green')
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()

# Pairplot to check relationships
sns.pairplot(df[numeric_columns], diag_kind='kde')
plt.show()
```

### DISTRIBUTION OF CATEGORICAL VARIABLES

```python

# List of categorical columns
categorical_columns = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area']

# Bar Charts
plt.figure(figsize=(10, 6))
for i, col in enumerate(categorical_columns, 1):
    plt.subplot(2, 3, i)
    sns.countplot(x=df[col], palette="viridis")
    plt.title(f'{col} Count')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Conclusion of the Loan Approval EDA Project

Based on the exploratory data analysis (EDA) conducted on the home loan approval dataset, the following insights were derived:

Data Quality and Preprocessing:

Some missing values were identified in the dataset. Handling these missing values (either by imputation or removal) is essential.
There were duplicate records in the dataset, which should be removed for better accuracy in analysis.
Numerical Feature Analysis:

The distribution of numerical features showed skewness in certain variables, indicating the presence of extreme values.
Boxplots revealed the presence of outliers in some financial variables, which may need further investigation.
Categorical Feature Analysis:

Some categorical variables had an imbalanced distribution, meaning certain loan categories were significantly more frequent than others.
Bar charts helped visualize the most common categories, providing insights into loan approval trends.
Correlation Analysis:

The correlation heatmap indicated that certain numerical variables had strong relationships. For example, income and loan amount might be positively correlated.
Weak correlations among other variables suggest that additional external factors might influence loan approvals.
Potential Next Steps:

Feature engineering could help improve predictive modeling by creating new meaningful variables.
Outliers could be handled using techniques like transformation or capping.
Advanced modeling techniques, such as machine learning algorithms, could be applied to predict loan approval outcomes.


# Final Thoughts:
The analysis provides an overview of the trends and relationships within the dataset. It sets a strong foundation for further predictive modeling or business decision-making regarding loan approvals.
